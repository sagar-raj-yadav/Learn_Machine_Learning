```python
import pandas as pd
```

```python
# I have multiple columns and mujhe kisi bhi column ka koi bhi row change karna h
df=pd.read_csv("data.csv")

print("before changing:-")
print(df)

print("after changing:-")
# jiska city 'B' hai usko 'D' krna h
for i in range(len(df)):
    if df.loc[i, 'city'] == 'B':
        df.loc[i, 'city'] = 'D'

print(df)


```

```python
# jiska city 'D' hai uska name 'akash' kar do

df.loc[df['city'] == 'D', 'Name'] = 'akash'

# Show updated DataFrame
print(df)
```

```python
"""
You want to fill missing values in your CSV where:
Numerical columns (Age) â†’ fill with average (mean)
Categorical columns (Name, city) â†’ fill with most frequent value (mode)
"""

def handle_missing_values(data):
    for col in data.columns:
        if data[col].dtype in ['int64', 'float64']:
            # Numerical column: fill with mean
            mean_value  = df[col].mean()
            df[col].fillna(mean_value)
        else:
            # Categorical column: fill with mode or "None"
            mode_value = df[col].mode()[0] if not df[col].mode().empty else "None"
            df[col].fillna(mode_value)
    return df

handle_missing_values(df)

print(df)
```


ðŸŸ¦ 1. Label Encoding
I. Converts categorical values into numeric labels (0, 1, 2, â€¦).
II. Each category gets a unique integer.
III. Output is a single column with numeric values.
IV. Used for ordinal data (where order matters, e.g., Low < Medium < High).
V. May cause issues for nominal data because the model might assume an order.
VI. Example:
Color â†’ Red = 0, Blue = 1, Green = 2
Education Level â†’ Primary = 0, Secondary = 1, Graduate = 2

=> Use Label Encoding when categories have order.


ðŸŸ© 2. One Hot Encoding

I. Converts categorical values into binary vectors (0/1).
II. Each unique category becomes a separate column.
III. The category present is marked as 1, others as 0.
IV. Used for nominal data (where no order exists, e.g., color, gender).
V. Avoids false order but increases the number of columns (dimensionality).
VI. Example:
Color â†’
Red = [1, 0, 0]
Blue = [0, 1, 0]
Green = [0, 0, 1]
Gender â†’ Male = [1, 0], Female = [0, 1]


=> Use One Hot Encoding when categories have no order.
