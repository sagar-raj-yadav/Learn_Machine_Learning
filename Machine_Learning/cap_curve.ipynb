```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
population_percentage = np.linspace(0, 100, 100)

# Perfect model: captures all positives immediately (steep curve)
perfect_model = np.piecewise(population_percentage, [population_percentage < 10, population_percentage >= 10],
                             [lambda x: 100*x/10, 100])

# Random model: diagonal line (random guessing)
random_model = population_percentage

# Example model: lies between perfect and random
example_model = np.piecewise(population_percentage, 
                             [population_percentage < 10, 
                              (population_percentage >= 10) & (population_percentage < 30), 
                              population_percentage >= 30],
                             [lambda x: 8*x, 
                              lambda x: 80 + 0.5*(x-10)*100/70, 
                              100])

# Plotting the CAP Curve
plt.figure(figsize=(8,6))
plt.plot(population_percentage, perfect_model, label='Perfect Model', color='green', linewidth=2)
plt.plot(population_percentage, random_model, label='Random Model', color='blue', linestyle='--', linewidth=2)
plt.plot(population_percentage, example_model, label='Your Model', color='red', linewidth=2)
plt.title('CAP Curve (Cumulative Accuracy Profile)')
plt.xlabel('Percentage of Population')
plt.ylabel('Percentage of Positives Captured')
plt.legend()
plt.grid(True)
plt.show()
```
